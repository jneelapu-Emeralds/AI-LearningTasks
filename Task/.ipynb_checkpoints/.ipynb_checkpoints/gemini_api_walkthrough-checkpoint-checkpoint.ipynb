```python
from PIL import Image
import matplotlib.pyplot as plt

cats_img = Image.open("images/image_2.png")

# Display the image using Matplotlib
plt.figure(figsize=(6, 4))
plt.imshow(cats_img)
plt.axis('off')  # Hide the axes
plt.show()
```

```python
# Import pipeline
from transformers import pipeline

# Specify the inference task
object_detection = pipeline(
    task="zero-shot-object-detection",
    model="google/owlvit-base-patch32"
  
)
```

```python
# Define candidate labels
candidate_labels=["squirrel", "nut", "wooden box"]

# Run detection
detections = object_detection(cats_img, candidate_labels=candidate_labels)
```

```python
from transformers import pipeline
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt



# Create drawing context
draw = ImageDraw.Draw(cats_img)
font = ImageFont.load_default()

# Draw detections
for detection in detections:
    box = detection['box']
    label = detection['label']
    score = detection['score']

    xmin, ymin, xmax, ymax = map(int, [
        box['xmin'], box['ymin'], box['xmax'], box['ymax']
    ])

    # Draw bounding box
    draw.rectangle([(xmin, ymin), (xmax, ymax)],
                   outline="red", width=4)

    text = f"{label}: {score:.2f}"

    # Get text size
    text_bbox = draw.textbbox((xmin, ymin), text, font=font)
    text_width = text_bbox[2] - text_bbox[0]
    text_height = text_bbox[3] - text_bbox[1]

    # Draw label background
    draw.rectangle(
        [(xmin, ymin - text_height),
         (xmin + text_width, ymin)],
        fill="red"
    )

    # Draw label text
    draw.text(
        (xmin, ymin - text_height),
        text,
        fill="white",
        font=font
    )

# Show image
plt.figure(figsize=(10, 6))
plt.imshow(cats_img)
plt.axis("off")
plt.show()
print("Total objects detected:", len(detections))
```

```python
from collections import Counter

# 1 Extract labels from detections
labels = [d["label"] for d in detections]

# 2 Count occurrences of each object
object_counts = Counter(labels)

# 3 Convert counts into natural language text
def counts_to_text(object_counts):
    sentences = []

    for label, count in object_counts.items():
        if count == 1:
            sentences.append(f"There is 1 {label} in the image.")
        else:
            sentences.append(f"There are {count} {label}s in the image.")

    return " ".join(sentences)

# Generate description
description = counts_to_text(object_counts)

print(description)
```

```python
labels = [d["label"] for d in detections if d["score"] > 0.3]

from collections import Counter

object_counts = Counter(labels)

def counts_to_text(object_counts):
    if not object_counts:
        return "No objects were detected in the image."

    parts = []

    for label, count in object_counts.items():
        if count == 1:
            parts.append(f"1 {label}")
        else:
            parts.append(f"{count} {label}s")

    if len(parts) == 1:
        return f"There is {parts[0]} in the image."
    else:
        return "There are " + ", ".join(parts[:-1]) + f", and {parts[-1]} in the image."


description = counts_to_text(object_counts)
print(description)
```

```python
from transformers import pipeline

narrator = pipeline(
    task="text-to-speech",
    model="suno/bark-small",
)
```

```python
narrated_text = narrator(description)
```

```python
print(narrated_text["sampling_rate"])
```

```python
import numpy as np
from IPython.display import Audio

audio = narrated_text["audio"]

# Convert to numpy if tensor
if hasattr(audio, "numpy"):
    audio = audio.numpy()

# Remove extra dimensions
audio = np.squeeze(audio)

Audio(audio, rate=narrated_text["sampling_rate"])
```
